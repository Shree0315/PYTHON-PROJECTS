# Extracting Clean Data

### Title of Website

```python
soup.title
```

### Price of the Product

```python
soup.find_all('div',class_='_30jeq3 _16Jk6d')
```

```python
price=soup.find_all('div',class_='_30jeq3 _16Jk6d')
price
```

```python
for i in price:
    print(i)
```

```python
len(price)
```

```python
product_price=[]

for i in range(0,len(price)):
    product_price.append(price[i].get_text())
```

```python
product_price[0]
```

### Get the names of the customers

```python
names=soup.find_all('p',class_='_2sc7ZR _2V5EHH')
names
```

```python
customer_names=[]

for i in range(0,len(names)):
    customer_names.append(names[i].get_text())
```

```python
customer_names
```

### Get the comments

```python
comments=soup.find_all('p',class_='_2-N8zT')
comments
```

```python
customer_comments=[]

for i in range(0,len(comments)):
    customer_comments.append(comments[i].get_text())
```

```python
customer_comments
```

### Get stars

```python
stars=soup.find_all('div',class_='_3LWZlK _1BLPMq')
stars
```

```python
customer_stars=[]

for i in range(0,len(stars)):
    customer_stars.append(stars[i].get_text())
```

```python
customer_stars
```

### Get the reviews

```python
reviews=soup.find_all('div',class_='t-ZTKy')
reviews
```

```python
customer_reviews=[]

for i in range(0,len(reviews)):
    customer_reviews.append(reviews[i].get_text())
```

```python
customer_reviews
```
